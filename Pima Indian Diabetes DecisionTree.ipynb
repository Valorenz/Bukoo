```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
df
```

```python
df.isnull().sum()
```

```python
X = df.drop(columns=['Outcome'])
Y = df['Outcome']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_test
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, Y_train)

model.get_params()
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

print('Accuracy: ', accuracy_score(Y_test, y_pred))
print('Recall: ', recall_score(Y_test, y_pred))
print('Precision: ', precision_score(Y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

print('Confusion matrix')
matrix = confusion_matrix(Y_test, y_pred)
print(matrix)
```

```python
import matplotlib.pyplot as plt

titles_options = [
    ("Confusion Matrix", None), ("Confusion Matrix Normalize", 'true')
]
for title, normalize in titles_options:
  disp = ConfusionMatrixDisplay.from_estimator(
      model,
      X=X_test,
      y=Y_test,
      cmap=plt.cm.Blues,
      normalize=normalize
  )
  disp.ax_.set_title(title)
```

# Parameter Tuning


```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [3, 5, 6, 7, 10, 15],
    'min_samples_leaf': [1, 2, 5, 7, 10, 15],
    'criterion':['gini', 'entropy']
}
```

```python
model = DecisionTreeClassifier(
    ccp_alpha=0.01,
    random_state=42
)
```

```python
grid = GridSearchCV(model, param_grid, cv=10, scoring='f1')
grid.fit(X_train, Y_train)
```

```python
new_model = grid.best_estimator_
y_pred = new_model.predict(X_test)

new_model.get_params()
```

```python
print('Accuracy: ', accuracy_score(Y_test, y_pred))
print('Recall: ', recall_score(Y_test, y_pred))
print('Precision: ', precision_score(Y_test, y_pred))
```

```python
titles_options = [
    ("Confusion Matrix", None), ("Confusion Matrix Normalize", 'true')
]

for title, normalize in titles_options:
  disp = ConfusionMatrixDisplay.from_estimator(
      new_model,
      X=X_test,
      y=Y_test,
      cmap=plt.cm.Blues,
      normalize=normalize
  )
  disp.ax_.set_title(title)
```

### Effect of `min_samples_split=3` and `min_samples_leaf=5` on the Model's Performance

During your hyperparameter tuning with `GridSearchCV`, the `new_model` was found to have optimal parameters including `min_samples_split=3` and `min_samples_leaf=5` (along with `criterion='entropy'` and `ccp_alpha=0.01`).

Let's break down the impact of these specific settings:

1.  **`min_samples_split=3`**: This setting dictates that an internal node must have at least 3 samples to be considered for splitting. This is a relatively small number, meaning the tree can still make splits on fairly small groups of data. It allows the model to capture more nuanced patterns than a much higher value would, but it's not so low (e.g., 2, which is the default) that it would easily lead to extreme overfitting by splitting on every single sample variation.

2.  **`min_samples_leaf=5`**: This setting ensures that each leaf node in the tree must contain at least 5 samples. This is a more restrictive condition than `min_samples_split=3`. It prevents the tree from creating leaf nodes that are too specific to individual data points or very small, noisy groups. By forcing a minimum of 5 samples per leaf, the model becomes more generalized, as each prediction made by a leaf is based on a slightly larger, more representative group of training examples. This is a key factor in reducing overfitting.

### Combined Effect and Performance Improvement

When combined with other tuned parameters like `criterion='entropy'` and `ccp_alpha=0.01`, these settings (`min_samples_split=3` and `min_samples_leaf=5`) led to a more optimized and robust Decision Tree model.

Let's compare the performance metrics:

*   **Initial Model (before tuning)**:
    *   Accuracy: 0.7056
    *   Recall: 0.7125
    *   Precision: 0.5588

*   **`new_model` (after tuning with `min_samples_split=3`, `min_samples_leaf=5`, etc.)**:

    *   Precision: 0.5842 (an increase of ~2.54%)

The improvements in accuracy, recall, and precision demonstrate that these specific parameter values (found through `GridSearchCV`) effectively balanced the model's complexity. They allowed the tree to learn meaningful patterns from the training data without becoming overly complex and susceptible to noise, leading to better generalization and improved performance on the unseen test data. The `min_samples_leaf=5` parameter likely played a significant role in making the leaf nodes more robust and less prone to capturing spurious relationships.

### `min_samples_split`

**Definition**: `min_samples_split` is the minimum number of samples required to split an internal node.

**Effect on the model**:
*   **Higher value**: A higher `min_samples_split` value means that a node must contain at least that many samples before it can be considered for splitting. This will prevent the tree from making splits on very small groups of samples, leading to a less complex tree. This can help to **reduce overfitting** by creating a more generalized model.
*   **Lower value**: A lower `min_samples_split` allows the tree to make more splits, even on smaller groups of samples. This can lead to a more complex tree that might capture more specific patterns in the training data, potentially leading to **overfitting** if the value is too low.

### `min_samples_leaf`

**Definition**: `min_samples_leaf` is the minimum number of samples required to be at a leaf node. A split point will only be considered if it leaves at least `min_samples_leaf` training samples in each of the left and right branches.

**Effect on the model**:
*   **Higher value**: A higher `min_samples_leaf` value forces each leaf node to contain a larger number of samples. This means the tree will be less granular and less likely to learn from noise in the data, thereby **reducing overfitting** and making the model more robust.
*   **Lower value**: A lower `min_samples_leaf` allows for leaf nodes with very few samples. This can lead to a very specific tree that might perfectly classify individual training examples but fail to generalize to new data, increasing the risk of **overfitting**.

### Combined Effect on the Model

Both `min_samples_split` and `min_samples_leaf` are crucial **regularization parameters** for Decision Trees. They control the complexity of the tree, which directly impacts the **bias-variance trade-off**:

*   **Preventing Overfitting**: By increasing these values, you restrict the tree from growing too deep and complex. This helps in **reducing variance** (making the model less sensitive to fluctuations in the training data) and thus preventing overfitting. However, setting them too high might lead to **underfitting** (high bias), where the model is too simple to capture the underlying patterns.

*   **Balancing Model Complexity**: The goal is to find an optimal balance. You want a tree that is complex enough to capture the important relationships in the data but simple enough to generalize well to unseen data. This is typically achieved through techniques like cross-validation and hyperparameter tuning (as demonstrated in your `GridSearchCV` step), where different combinations of these parameters are tested to find the best performing model.

```python
import pandas as pd
from sklearn.model_selection import cross_validate
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score



X = df.drop('Outcome', axis=1)
y = df['Outcome']


dt_model = DecisionTreeClassifier(random_state=42)

# 4. SKENARIO PENGUJIAN (CROSS-VALIDATION)
# Menyiapkan metrik evaluasi yang diminta
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

cv_results = cross_validate(dt_model, X, y, cv=5, scoring=scoring)

# 5. MENAMPILKAN HASIL EVALUASI
# Membuat DataFrame untuk tabel hasil per fold
results_df = pd.DataFrame({
    'Fold': range(1, 6),
    'Akurasi': cv_results['test_accuracy'],
    'Precision': cv_results['test_precision'],
    'Recall': cv_results['test_recall'],
    'F1-Score': cv_results['test_f1']
})

print("\n=== Tabel 2. Hasil Evaluasi Data (Per Fold) ===")
print(results_df.round(3))

# Menghitung dan menampilkan rata-rata
print("\n=== Kesimpulan Rata-rata Evaluasi ===")
print(f"Rata-rata Akurasi  : {results_df['Akurasi'].mean():.3f}")
print(f"Rata-rata Precision: {results_df['Precision'].mean():.3f}")
print(f"Rata-rata Recall   : {results_df['Recall'].mean():.3f}")
print(f"Rata-rata F1-Score : {results_df['F1-Score'].mean():.3f}")
```
